```r
# We then pass df to the prcomp(). We also set two arguments, center and scale, 
# to be TRUE then preview our object with summary

mtcars.pca <- prcomp(mtcars[,c(1:7,10,11)], center = TRUE, scale. = TRUE)
summary(mtcars.pca)
```

```r
#Extract the std from PCA result to calculate the variance

var= mtcars.pca$sdev^2
(pve= var/sum(var))
cumsum(pve)
```

## IRIS DATASET PERFORMING PCA AND VISUALIZATIONS

```r
head(iris)
```

```r
library('')
```

```r
dim(iris)
```

```r
#Perform PCA on the four continous variables 
#Apply log transformation and scale

log.iris= log(iris[,1:4])
iris.species= iris[, 5]

#Apply PCA and Scale

iris.pca= prcomp(log.iris,
                center= TRUE,
                scale. = TRUE)
```

```r
#Analyzing the result

print(iris.pca)
```

```r
#Plot method

plot(iris.pca, type = 'l')
```

The above plot returns a plot of the variances(y-axis) associated with the PC's (x-axis).

It is useful to decide how many PC's to retain for further analysis.In this case only the first 2 explain the mostnof the variability in the data

```r
#Summarize the important PCA

summary(iris.pca)
```

> GRAPH OF INDIVIDUALS

```r
#Graph of individuals, individuals with a similar profile are grouped together

fviz_pca_ind(iris.pca,
            col.ind= "cos2", #Color by the quality of representation
            gradient.cols= c("#00AFBB", "#E7B800", "#FC4E07"),
            repel= TRUE  #Avoid text overlapping
            )
```

> GRAPH OF VARIABLES

```r
#Positive correlated variables point to the same side of the plot
#Negative correlated variables  point to the opposite side of the graph

fviz_pca_var(iris.pca, col.var= "contrib" , # Color by contributions to the PC
            gradient.cols= c("#00AFBB", "#E7B800", "#FC4E07"),
            repel= TRUE #Avoid overlapping
            )
```

> ACCESS TO THE PCA RESULT

```r
#Eigenvalues

eig.val= get_eigenvalue(iris.pca)
eig.val
```

```r
#Result for variables

res.var= get_eigenvalue(iris.pca)
res.var$coord #Coordinates
res.var$contrib #Contributions to the PCs
res.var$cos2  #Quality of representation
```

```r
#Results for individuals

res.ind = get_pca_ind(iris.pca)
res.ind$coord
res.ind$contrib
res.ind$cos2
```

```r
#Plotting

theta = seq(0,2*pi, length.out = 100)
circle = data.frame(x = cos(theta), y = sin(theta))
p= ggplot(circle, aes(x,y)) + geom_path()

loadings = data.frame(iris.pca$rotation,
                     .names= row.names(iris.pca$rotation))

p + geom_text(data = loadings,
             mapping= aes(x= PC1, y=PC2, label= .names, colour= .names))+
coord_fixed(ratio= 1)+
labs(x= "PC1", y="PC2")
```
